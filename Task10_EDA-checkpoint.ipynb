```python
##adding outliers##
numeric_cols = ["sales", "profit", "quantity", "discount"]
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers, lower_bound, upper_bound
```

```python
for col in numeric_cols:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    
    print(f"\nColumn: {col}")
    print(f"Lower Bound: {lower}")
    print(f"Upper Bound: {upper}")
    print(f"Number of Outliers: {outliers.shape[0]}")
```

```python
##creating outlier flag coloumn## 
def add_outlier_flag_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    flag_col = f"{column}_outlier_flag"
    df[flag_col] = ((df[column] < lower_bound) | (df[column] > upper_bound)).astype(int)
    
    return df
```

```python
for col in numeric_cols:
    df = add_outlier_flag_iqr(df, col)
```

```python
df[[col + "_outlier_flag" for col in numeric_cols]].sum()
```

```python
###to cap outliers##
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df
```

```python
# Cap these columns(selective capping)
for col in ["sales", "quantity"]:
    df = cap_outliers_iqr(df, col)

# Cap discount explicitly (business rule)
df["discount"] = df["discount"].clip(0, 0.50)

# Profit: keep as-is (do NOT cap or remove)
```
